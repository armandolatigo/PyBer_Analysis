```python
#pie charts
import matplotlib.pyplot as plt

categories = ['Jurassic Park', 'The Lost World: JP 2', 'Jurassic Park III', 'Jurassic World', 'Jurassic World: Fallen Kingdom']
rating = [97, 91, 52, 88, 64]

plt.pie(rating)
plt.show()
```

```python
explode = [0.2,0,0,0,0]
plt.pie(rating,labels=categories, autopct='%.1f%%',explode=explode)
plt.show()
```
